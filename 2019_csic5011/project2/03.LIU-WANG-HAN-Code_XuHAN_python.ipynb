```python
import numpy as np
from scipy.io import loadmat
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.manifold import MDS, Isomap, LocallyLinearEmbedding, TSNE, SpectralEmbedding
```

```python
plt.rcParams.update({'font.size': 22})
```

```python
mat = loadmat('face.mat')
X = np.reshape(mat['Y'],(10304,33)).T/255
ID = mat['id']
```

```python
mat['id']
```

```python
def emb1(f,title):
    embedding = f(n_components=2)
    X_transform = embedding.fit_transform(X)
    plt.scatter(X_transform[:,0],X_transform[:,1])
    print(np.argsort(X_transform[:,0])+1)
    plt.title(title)
    
def emb2(f,k,title):
    embedding = f(n_neighbors=k, n_components=2)
    X_transform = embedding.fit_transform(X)
    plt.scatter(X_transform[:,0],X_transform[:,1])
    print(np.argsort(X_transform[:,0])+1)
    plt.title(title)
    
def emb3(f,k,method,title):
    embedding = f(n_neighbors=k, method=method, n_components=2)
    X_transform = embedding.fit_transform(X)
    plt.scatter(X_transform[:,0],X_transform[:,1])
    print(np.argsort(X_transform[:,0])+1)
    plt.title(title)
```

MDS

```python
emb1(MDS,'MDS')
```

ISOMAP

```python
emb2(Isomap,5,'ISOMAP')
```

LLE

```python
emb3(LocallyLinearEmbedding,5,'standard','LLE')
```

LTSA

```python
emb3(LocallyLinearEmbedding,5,'ltsa','LTSA')
```

t-SNE

```python
emb1(TSNE,'t-SNE')
```
